```python
import pandas as pd
import random
import numpy as np

# Load test.csv
test_data = pd.read_csv("/kaggle/input/llms-you-cant-please-them-all/test.csv")
```

```python

# List of sample topics (you would replace this with your actual dataset of topics)
topics = [
    "The impact of technology on society",
    "Should schools enforce dress codes?",
    "The role of artificial intelligence in modern education",
    "Climate change: myth or reality?",
    "The influence of social media on mental health",
    "The ethics of genetic engineering",
    "Can democracy survive in the age of misinformation?"
]

# Essay generation tactics
def generate_essay(topic):
    """
    Generate an essay based on the given topic.
    Use strategies to introduce ambiguity, complexity, and contradictions.
    """
    # strategies for essay generation:
    # 1. Contradictory statements
    # 2. Ambiguity (e.g., unanswered questions)
    # 3. Mixed tones (formal and informal)
    # 4. Unconventional grammar

    # Strategy 1: Contradictory statements or ambiguous conclusions
    if topic == "The impact of technology on society":
        essay = """
        Technology has undoubtedly improved our lives in many ways. From healthcare advancements to better communication tools, it‚Äôs clear that we‚Äôre moving toward a future shaped by tech. However, as we rely more on these technologies, some argue we‚Äôre losing the very things that make us human. 
        Is technology bringing us closer together or pushing us further apart? One could say it's both.
        """
    
    elif topic == "Should schools enforce dress codes?":
        essay = """
        Schools should enforce dress codes to maintain discipline and a professional atmosphere. However, some believe that dress codes infringe on personal freedom. On the other hand, uniforms can actually reduce peer pressure and bullying. It's an interesting debate ‚Äì one that doesn't have a clear answer.
        """
    
    elif topic == "The role of artificial intelligence in modern education":
        essay = """
        AI in education promises personalized learning experiences and more efficient administrative tasks. Some argue that AI could replace teachers in the future. But others worry that relying too much on AI may limit the human touch that is necessary in education. It's a complex situation.
        """
    
    elif topic == "Climate change: myth or reality?":
        essay = """
        Climate change is a pressing global issue, and while the majority of scientists agree on its existence, there are still voices of dissent. Some question the evidence behind global warming, while others claim the data is undeniable. If climate change is real, what should we do about it? The solutions are still debated.
        """
    
    elif topic == "The influence of social media on mental health":
        essay = """
        Social media can be harmful to mental health, fostering feelings of loneliness and inadequacy. However, it can also connect people and help raise awareness of important causes. Some people argue that social media isn‚Äôt the problem ‚Äì it‚Äôs how we use it. What is the true impact? It depends on who you ask.
        """
    
    elif topic == "The ethics of genetic engineering":
        essay = """
        Genetic engineering holds incredible promise, from curing diseases to creating more resilient crops. But where should we draw the line? Editing the human genome raises serious ethical questions. Some believe we should proceed with caution, while others argue that it could be humanity's next evolutionary step.
        """
    
    elif topic == "Can democracy survive in the age of misinformation?":
        essay = """
        Democracy is based on the free exchange of ideas, but in the age of misinformation, it‚Äôs becoming increasingly difficult to separate fact from fiction. While some argue that democracy can adapt and find ways to combat fake news, others worry that we may be losing the very foundation of our democratic process.
        """
    
    return essay.strip()

# Function to simulate submission creation
def create_submission_file(topics):
    submission = []
    for i, topic in enumerate(topics, 1):
        essay = generate_essay(topic)
        submission.append([i, essay])
    # Create a DataFrame for easy export to CSV
    submission_df = pd.DataFrame(submission, columns=['id', 'essay'])
    return submission_df
```

```python
# Generate the submission file
submission_df = create_submission_file(topics)

# Save to CSV (make sure to replace 'sample_submission.csv' with your actual file path)
submission_df.to_csv('submission.csv', index=False)

# Display the first few rows to verify
print(submission_df.head())
```
