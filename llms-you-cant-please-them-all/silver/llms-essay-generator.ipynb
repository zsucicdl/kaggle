```python
import pandas as pd
import random
import numpy as np

# Load test.csv
test_data = pd.read_csv("/kaggle/input/llms-you-cant-please-them-all/test.csv")
```

```python
test_data.head()
```

```python

# Templates with confusion elements
INTRO_TEMPLATES = [
    "{topic} is both a solution and a problem, depending on how one chooses to interpret it.",
    "While {topic} seems straightforward, its implications unravel a tapestry of complexities.",
    "Discussing {topic} feels like peeling an onion‚Äîeach layer reveals something new, yet the core remains elusive.",
]

FOR_TEMPLATES = [
    "Proponents believe that {topic} is the key to progress, yet some of their arguments seem to contradict their goals. For example, fostering growth while maintaining control feels inherently paradoxical.",
    "{topic} opens pathways to opportunities, but do these opportunities lead to real progress or just a mirage of advancement?",
    "Supporters argue that {topic} enables progress, yet the very nature of progress is subjective and possibly misleading.",
]

AGAINST_TEMPLATES = [
    "Critics argue that {topic} poses challenges, but are these challenges genuine obstacles or distractions from deeper issues?",
    "While {topic} has its critics, some of their concerns feel more like fears of change rather than substantial critiques.",
    "Opponents caution against {topic}, yet their resistance often raises more questions than it answers. What are we truly resisting?",
]

CONCLUSION_TEMPLATES = [
    "In conclusion, {topic} is like a coin with infinite sides; we can analyze it forever, yet its true nature may remain hidden.",
    "Ultimately, the discourse around {topic} reveals more about us than about the topic itself‚Äîare we capable of clear judgment, or do we thrive in ambiguity?",
    "Perhaps the real question isn‚Äôt whether {topic} is good or bad, but whether we understand it at all.",
]

# Function to generate a deliberately ambiguous essay
def generate_ambiguous_essay(topic):
    # Apply lower() only where necessary
    topic_lower = topic.lower()
    
    introduction = random.choice(INTRO_TEMPLATES).format(topic=topic)
    arguments_for = random.choice(FOR_TEMPLATES).format(topic=topic_lower)
    arguments_against = random.choice(AGAINST_TEMPLATES).format(topic=topic_lower)
    conclusion = random.choice(CONCLUSION_TEMPLATES).format(topic=topic_lower)

    essay = f"{introduction}\n\n{arguments_for}\n\n{arguments_against}\n\n{conclusion}"
    return essay

# Function to create the submission file
def create_submission_file(test_data):
    essays = []
    for _, row in test_data.iterrows():
        topic = row['topic']
        essay = generate_ambiguous_essay(topic)
        essays.append({'id': row['id'], 'essay': essay})

    submission_df = pd.DataFrame(essays)
    return submission_df
```

```python
# Generate the submission file
submission_df = create_submission_file(test_data)

# Save to CSV
submission_df.to_csv('submission.csv', index=False)

# Display the first few rows to verify
submission_df.head()
```
