```python
%%writefile config.yml
task: text_regression
base_model: google-bert/bert-base-uncased # choose any model from huggingface hub!!!
project_name: autotrain-essay-scoring-model
log: tensorboard
backend: local-cli

data:
  path: /kaggle/input/learning-agency-lab-automated-essay-scoring-2
  train_split: train # this must be either train.csv or train.json
  valid_split: null # this must be either valid.csv or valid.json, null for auto-split
  column_mapping:
    text_column: full_text # this must be the name of the column containing the text
    target_column: score # this must be the name of the column containing the target

params:
  max_seq_length: 512
  epochs: 1
  batch_size: 4
  lr: 2e-5
  optimizer: adamw_torch
  scheduler: linear
  gradient_accumulation: 1
  mixed_precision: fp16

hub:
  username: ${HF_USERNAME}
  token: ${HF_TOKEN}
  push_to_hub: true
```
