**TLDR; 33 cols can be removed!**

```python
from itertools import combinations
import pandas as pd
from tqdm.auto import tqdm
```

```python
train = pd.read_csv('/kaggle/input/um-game-playing-strength-of-mcts-variants/train.csv')

agg = train.nunique()
cols_to_drop = agg[agg<=1].index

train.drop(cols_to_drop, axis=1, inplace=True)
```

```python
# Game features only for faster compute: not dropping gives same result for this dataset
train = train.drop_duplicates('GameRulesetName')
```

```python
column_pairs = [(i, j) for i,j in tqdm(combinations(train, 2)) if train[i].equals(train[j])]
```

```python
canonical = {}
for c1, c2 in column_pairs:
    if c1 in canonical:
        canonical[c1].append(c2)
    else:
        canonical[c1] = [c2]
```

```python
canonical
```

```python
repeated_cols = set([c2 for _, c2 in column_pairs]) # c1 will be the ones we keep
```

```python
print('Number of cols to be removed:', len(repeated_cols))
```

```python
print(list(sorted(repeated_cols)))
```

```python
for c in sorted(repeated_cols):
    print(c)
```

```python
with open('repeated_cols.txt','w') as f:
    for c in repeated_cols:
        f.write(c + '\n')
```

```python
# sanity check
train.drop(repeated_cols,axis=1,inplace=True)
column_pairs = [(i, j) for i,j in tqdm(combinations(train, 2)) if train[i].equals(train[j])]
assert len(column_pairs) == 0
```
