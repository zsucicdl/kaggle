```python
import pandas as pd
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(20, 15))

for i in range(5):
    for j in range(5):
        start_row = i * 50
        end_row = start_row + 50
        start_col = j * 200
        end_col = start_col + 200
        
        axes[i, j].matshow(train_100468857_fgs1.iloc[start_row:end_row, start_col:end_col])
        axes[i, j].set_title(f"Rows {start_row}-{end_row}, Columns {start_col}-{end_col}")

plt.tight_layout()
plt.show()
```
