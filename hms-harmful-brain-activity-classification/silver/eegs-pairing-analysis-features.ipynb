```python
import numpy as np
import pandas as pd
import seaborn as sns
from tqdm import tqdm
import matplotlib.pyplot as plt
from IPython.display import Image, display, HTML
sample = pd.read_parquet("/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/1000913311.parquet")
train = pd.read_csv("/kaggle/input/hms-harmful-brain-activity-classification/train.csv")
```

```python
import pywt
def maddest(d, axis=None):
    return np.mean(np.absolute(d - np.mean(d, axis)), axis)

def denoise(x, wavelet='haar', level=1):
    ret = {key:[] for key in x.columns}
    
    for pos in x.columns:
        coeff = pywt.wavedec(x[pos], wavelet, mode="per")
        sigma = (1/0.6745) * maddest(coeff[-level])

        uthresh = sigma * np.sqrt(2*np.log(len(x)))
        coeff[1:] = (pywt.threshold(i, value=uthresh, mode='hard') for i in coeff[1:])

        ret[pos]=pywt.waverec(coeff, wavelet, mode='per')
    
    return pd.DataFrame(ret)
```

```python
def plot_comparision(data):
    data = denoise(data, wavelet='db8')
    plt.figure(figsize=(15, 45))
    for i, (channel1, channel2) in enumerate(pairing.items()):
        plt.subplot(len(pairing), 1, i + 1)
        sns.lineplot(data=data[[channel1, channel2]], alpha=0.5)
        plt.title(f"Comparison of {channel1} and {channel2}")
        plt.xlabel("Time")
        plt.ylabel("Values")
        plt.legend([channel1, channel2])

    plt.tight_layout()
    plt.show()
plot_comparision(sample)
```

```python
def plot_difference(data):
    data = denoise(data, wavelet='db8')
    plt.figure(figsize=(15, 45))
    for i, (channel1, channel2) in enumerate(pairing.items()):
        plt.subplot(len(pairing), 1, i + 1)
        difference = data[channel1] - data[channel2]
        sns.lineplot(data=difference)
        plt.title(f"Difference between {channel1} and {channel2}")
        plt.xlabel("Time")
        plt.ylabel("Difference in Values")
    plt.tight_layout()
    plt.show()
plot_difference(sample)
```

```python
df = pd.read_csv("/kaggle/input/hms-harmful-brain-activity-classification/train.csv")
TARGETS = [x for x in df.columns if 'vote' in x]
train = df.groupby('eeg_id')[['spectrogram_id','spectrogram_label_offset_seconds']].agg(
    {'spectrogram_id':'first','spectrogram_label_offset_seconds':'min'})
train.columns = ['spec_id','min']

tmp = df.groupby('eeg_id')[['spectrogram_id','spectrogram_label_offset_seconds']].agg(
    {'spectrogram_label_offset_seconds':'max'})
train['max'] = tmp

tmp = df.groupby('eeg_id')[['patient_id']].agg('first')
train['patient_id'] = tmp

tmp = df.groupby('eeg_id')[TARGETS].agg('sum')
for t in TARGETS:
    train[t] = tmp[t].values
    
y_data = train[TARGETS].values
y_data = y_data / y_data.sum(axis=1,keepdims=True)
train[TARGETS] = y_data

tmp = df.groupby('eeg_id')[['expert_consensus']].agg('first')
train['target'] = tmp

train = train.reset_index()
print('Train non-overlapp eeg_id shape:', train.shape )
train.head()
```
