```python
color_hash = {'Cz': 'khaki',
 'Fpz': 'lightgreen',
 'Fz': 'lightgreen',
 'Pz': 'lightgreen',
 'Oz': 'lightgreen',
 'T7': 'skyblue',
 'C3': 'skyblue',
 'C4': 'skyblue',
 'T8': 'skyblue',
 'Fp2': 'salmon',
 'F8': 'salmon',
 'P8': 'violet',
 'O2': 'violet',
 'Fp1': 'salmon',
 'F7': 'salmon',
 'P7': 'violet',
 'O1': 'violet',
 'F3': 'salmon',
 'F4': 'salmon',
 'P3': 'violet',
 'P4': 'violet'}
from eeg_positions import get_elec_coords, plot_coords
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
coords = get_elec_coords(
    system="1020",
    dim="2d",
)
chans = list(coords['label'])
def egg_index_colors(parse_chans):
    pos, colrs = [],[]
    for i, x in enumerate(chans):
        if x not in parse_chans + ['T7','T8','P7','P8']:
            pos.append(i)
            colrs.append('white')
        else:
            pos.append(i)
            colrs.append(color_hash[x])
    return pos, colrs
parse_chans = list(set(eegs.columns) - set([ 'EKG']))
egg_chans, egg_colors = egg_index_colors(parse_chans)
```

```python
from eeg_positions import get_elec_coords, plot_coords
coords = get_elec_coords(
    system="1020",
    dim="2d",
)
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
colors = (
    egg_colors
)

fig, ax = plot_coords(
    coords,
    scatter_kwargs={
        "s": 300, 
        "color": colors,
        "edgecolors": "black", 
        "linewidths": 0.5, 
    },
    text_kwargs={
        "ha": "center",  
        "va": "center",  
        "fontsize": 8, 
    },
)
```
