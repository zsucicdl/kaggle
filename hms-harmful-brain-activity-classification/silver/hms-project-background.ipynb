# Background



## What is EEG? 
*See ref 1 & 2*

Electroencephalography (EEG) uses non-invasive electrodes placed on the scalp and measures the current produced during excitation of groups of neurons. The weak electrical current that these groups of neurons produce is massively amplified, resulting in an EEG recording.

The waves that we see in an EEG recording are sinusoidal in nature, and they are normally measured from peak to trough. There are four main classifications for EEG brain wave patterns: 

1. Beta waves - more than 13Hz
2. Alpha waves - 8-13Hz
3. Theta waves - 4-8Hz
4. Delta waves - 0.5-4Hz

Depending on the state of the subject one waveform will be more dominant thatn the rest (ex: beta waves dominate during wakefulness, delta waves dominate during non-REM deep sleep). 


## Harmful EEG Patterns

*See ref 3*

Note: All EEG patterns will not be perfectly rhythmic and will have some background noise. The waves are determined by the predominant pattern. 

Arrythmic or polymorphic patterns mean that there is pathology there - just not associated with seizures.

### Helpful measurement terms 

<u>**Voltage**</u>

- Absolute voltage (A) is measured peak to trough
- Relative voltage (A/A<sub>b</sub>) is the voltage of discharge amplitude (A) compared with the background amplitude (A<sub>b</sub>)

<u>**Phase**</u>

- Area under the pattern curve on one side of the baseline (midline of the reading)

### Patterns associated with seizures

General Absence Seizures show a cortical-wide change in brain wave amplitude to ~3Hz continuously for at least 5 minutes until the seizure ends.

For any of these patterns to be a bona fide pattern, they need to occur at least 6 times one after another.

<u>__Periodic discharges__</u>

Periodic discharges with intervals between them

Features: 
- Phase (area under curve on one side of the baseline): ≤0.5s regardless of phases or >0.5s and have ≤3 phases

**Generalized Periodic Discharges (GPD)**: Across both hemispheres and electrodes, there is a time-locked pattern. Pathological GPDs often have long intervals between discharges. 

**Lateralised Periodic Discharges (LPD)**: Periodic discharges in one hemisphere of the brain. LPDs are associated with seizures but can also occur without seizures.

Note: Propagation from one side of the brain to the other would take the form of a time-locked discharge on both sides, not to be confused with LPDs (read below) that can be asynchronous.
 
<u>__Rhythmic Delta Activity__</u>

Discharges with no intervals. If not rhythmic nor periodic, these can be classified as arrythmic, irregular or polymorphic

**Generalised rhythmic delta activity (GRDA)**: Periodic discharges across all electrodes.

**Lateralised rhythmic delta activity (LRDA)**: One sided rhythmic delta waves. Very high association with seizures.

<u>__Highly epileptiform bursts__</u>: 

Bursts are patterns that are longer than 1 second (longer than PDs) and have more than 4 phases (PDs longer than 1 second must have less than 3 phases).

Within at least 50% of the bursts, there are there are 2 or more epileptiform discharges.

<u>**Spike and Wave (rare)**</u>

Tight spikes and a longer wave afterwards, with no intervals

<u>**Sporadic epileptiform discharges**</u>

Non-rhythmic and non-periodic epileptiform discharges that are associated with epilepsy, occurring before or after seizures.

### Features that make these patterns more "seizure-like"

Associated with more severe seizures and poorer patient outcomes.

<u>__Plus modifiers__</u>

Features affecting *amplitude*

- **+S** - means that there are frequent and/or sharp spikes superimposed (timelocked) onto a delta wave rhythm ONLY. Spikes dissipate once the RDA is over.
- **+F** - Fast frequency (theta or faster) superimposed (timelocked) onto RDAs and PDs. The fast frequencies dissipate once the RDA or PD is over. 
- **+R** - superimposed rhythmic delta activity (or a similar looking pattern) on a PD ONLY. Basically when PD and RDA occur at the same time.

<u>__Evolving or fluctuating modifiers__</u>

Features affecting *frequency, location, and/or morphology*. If absent, the pattern is static.

**Evolving**

Evolving patterns must have 2 changes in at least one of the following features:
- Frequency has ≥2 consecutive increases or decreases of more than 0.5 seconds (2--> 2.5 --> 3.5)
- Morphology has ≥2 consecutive changes to a novel morphology (ex: smooth to bumpy to spikey)
- The changes spread into or out of ≥2 different electrode locations (ex: same pattern moves or spreads from the frontal to parietal electrodes, or from one hemishpere to another).

For a seizure, pattern must be evolving for ≥ 10s, and reach > 4Hz. Otherwise class as 'other'. Note: you can have an evolving RDA without a seizure <10s, <4Hz. 

On Spectograms, the changes in frequency look like diagonals.

**Fluctuating** 

If not evolving, fluctuating must have at least 3 changes in one of the following characteristics: 
- ≥ 3 Frequency changes by ≥0.5/s (4Hz--> 2Hz--> 4Hz--> 2Hz)
- ≥ 3 Morphology changes
- ≥ 3 location changes 

<u>**Sharpness**</u>

Sharp 
- 70-200ms

Spiky
- <70ms 

Where the predominant phase has the greatest amplitude.
