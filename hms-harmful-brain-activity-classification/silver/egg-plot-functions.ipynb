```python
def plotegg(df, smooth=1):
    fig, ax = plt.subplots(20,sharex=True)
    for plnr, elect in zip(range(20),df.keys()):
        color = 'r' if elect == 'EKG' else 'k'
        ax[plnr].plot(df[elect].rolling(smooth).mean(),color)
        ax[plnr].grid()
        ax[plnr].set_ylabel(elect, rotation=45)
    plt.subplots_adjust(top=1.0,bottom=0.03,left=0.035,right=1.0,hspace=0.0,wspace=0.2)
    plt.xticks(np.arange(0, 10000, 200),np.arange(0,10000//200))
```

```python
plotegg(df,smooth=12)
```

```python
import mne
import pandas as pd
```

```python
def read_parquet_into_mne(path,freq=200):
    df = pd.read_parquet(path)
    ch_names = list(df.keys())
    df = df.T
    ch_types = (["eeg"] * (len(ch_names)-1)) + ['ecg']
    info = mne.create_info(ch_names, ch_types=ch_types, sfreq=freq)
    info.set_montage("standard_1020")
    raw = mne.io.RawArray(df, info, verbose=True)
    raw.info = info
    return raw
```

```python
raw = read_parquet_into_mne('/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/1000913311.parquet')
```
