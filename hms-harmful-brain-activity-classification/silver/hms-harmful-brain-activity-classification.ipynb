```python
directory_EEG = "/kaggle/input/hms-harmful-brain-activity-classification/train_eegs"
```

```python
all_files_EEG = os.listdir(directory_EEG)
```

```python
target_filename = "1000913311.parquet"

# Construct the full path
target_file_path = os.path.join(directory_EEG, target_filename)

# Check if the file exists
if os.path.exists(target_file_path):
    print(f"Found target file: {target_file_path}")
else:
    print(f"File '{target_filename}' not found in the directory.")
```

```python
len(all_files_EEG)
```

```python
columns = train_eeg_1.columns
num_plots = len(columns)
num_cols = 2
num_rows = (num_plots + num_cols - 1) // num_cols

fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, num_rows * 4))  

for i, col in enumerate(columns):
    ax = axes[i // num_cols, i % num_cols] if num_rows > 1 else axes[i % num_cols]
    ax.plot(train_eeg_1[col])
    ax.set_title(col + ' data')
    ax.set_xlabel('Index')
    ax.set_ylabel('Values')

plt.tight_layout(pad=3.0)  
plt.show()
```
