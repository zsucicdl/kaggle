# create features from eegs

```python
paths=[]
ids=[]
for dirname, _, filenames in os.walk('/kaggle/input/hms-harmful-brain-activity-classification/train_eegs'):
    for filename in filenames:
        path=os.path.join(dirname, filename)
        ids+=[filename.split('.')[0]]
        paths+=[path]
        
train_eggs=pd.DataFrame(columns=['eeg_id','path'])
train_eggs['path']=paths
train_eggs['eeg_id']=ids

for i in range(len(paths)):
    path=paths[i]
    table = pq.read_table(path)
    df = table.to_pandas()
    for column in columns_eegs:
        train_eggs.loc[i,column+'_min']=df[column].min()
        train_eggs.loc[i,column+'_max']=df[column].max()
#display(train_eggs)
train_eggs.to_csv('train_eggs.csv',index=False)
```

```python
paths=[]
ids=[]
for dirname, _, filenames in os.walk('/kaggle/input/hms-harmful-brain-activity-classification/test_eegs'):
    for filename in filenames:
        path=os.path.join(dirname, filename)
        ids+=[filename.split('.')[0]]
        paths+=[path]
        
test_eggs=pd.DataFrame(columns=['eeg_id','path'])
test_eggs['path']=paths
test_eggs['eeg_id']=ids

for i in range(len(paths)):
    path=paths[i]
    table = pq.read_table(path)
    df = table.to_pandas()
    for column in columns_eegs:
        test_eggs.loc[i,column+'_min']=df[column].min()
        test_eggs.loc[i,column+'_max']=df[column].max()
#display(test_eggs)
test_eggs.to_csv('test_eggs.csv',index=False)
```
