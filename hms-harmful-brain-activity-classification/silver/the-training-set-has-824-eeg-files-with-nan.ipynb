```python
%%time
df = pd.read_csv('/kaggle/input/hms-harmful-brain-activity-classification/train.csv')

nan_list = []
list_of_unique_eeg_id = df['eeg_id'].unique()
for eeg in list_of_unique_eeg_id:
    filepath = f'/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/{eeg}.parquet'
    matrix = pd.read_parquet(filepath).values
    if np.isnan(matrix).any():
        nan_list.append(eeg)
        #print(eeg)
        
print(f'There are {len(nan_list)} eegs containing at least a NaN.  ')
```

```python
print(nan_list[:3])
np.save('eeg_containing_nan.npy', nan_list)
```
