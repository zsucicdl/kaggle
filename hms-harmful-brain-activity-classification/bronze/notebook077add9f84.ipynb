```python
import pandas as pd

# Assuming 'predictions' is a DataFrame with columns 'id' and 'target'
# 'id' represents the unique identifier for each test instance, and 'target' is the predicted value

df1 = pd.read_csv("/kaggle/input/hms-harmful-brain-activity-classification/sample_submission.csv")
print(df1)

predictions =pd.concat([df1['eeg_id'],df1['seizure_vote'],df1['lpd_vote'],df1['gpd_vote'],df1['lrda_vote'],df1['grda_vote'],df1['other_vote']], axis=1)


# Create a submission  DataFrame
submission_df = pd.DataFrame({
    'eeg_id': predictions['eeg_id'],
    'seizure_vote':predictions['seizure_vote'],
    'lpd_vote':predictions['lpd_vote'],
    'gpd_vote': predictions['gpd_vote'],
    'lrda_vote': predictions['lrda_vote'],
    'grda_vote' : predictions['grda_vote'],
    'other_vote': predictions['other_vote']
})

# Save the submission DataFrame to a CSV file
submission_df.to_csv('submission.csv', index=False)
```
