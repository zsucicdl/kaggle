```python
import os
import pandas as pd, numpy as np
from tqdm import tqdm
import pickle
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/hms-harmful-brain-activity-classification/train.csv')
df.head()
```

From the dataset description, we can know:
* Each patient's data is split into several subsamples.
* Each subsample covers 50 seconds.

We can use the **'eeg_label_offset_seconds'** to split off each subsample.

```python
# here is an exmaple, patient_id=30631 has 270 eeg_ids, we pick one 2211351621.
df_example = df[(df.patient_id==30631)&(df.eeg_id==2211351621)].reset_index(drop=True)
display(df_example)
```

Here we can see **eeg_label_offset_seconds** includes [0, 6, 10, 34, 68, 78, 82].
Load its EEG data.

```python
eeg_example = pd.read_parquet('/kaggle/input/hms-harmful-brain-activity-classification/train_eegs/2211351621.parquet')
print(f'eeg_id=2211351621 has {eeg_example.shape[0]} samples.')
print()
display(eeg_example.head())
```

```python
# The range of eeg_sub_id=0
start_0, end_0 = 0*200, (0+50)*200
print(f"eeg_sub_id=0 range from {start_0} to {end_0}")

# The range of eeg_sub_id=1
start_1, end_1 = 6*200, (6+50)*200
print(f"eeg_sub_id=1 range from {start_1} to {end_1}")

# The range of eeg_sub_id=2
start_2, end_2 = 10*200, (10+50)*200
print(f"eeg_sub_id=2 range from {start_2} to {end_2}")

# The range of eeg_sub_id=3
start_3, end_3 = 34*200, (34+50)*200
print(f"eeg_sub_id=3 range from {start_3} to {end_3}")

# The range of eeg_sub_id=4
start_4, end_4 = 68*200, (68+50)*200
print(f"eeg_sub_id=4 range from {start_4} to {end_4}")

# The range of eeg_sub_id=5
start_5, end_5 = 78*200, (78+50)*200
print(f"eeg_sub_id=5 range from {start_5} to {end_5}")

# The range of eeg_sub_id=6
start_6, end_6 = 82*200, (82+50)*200
print(f"eeg_sub_id=6 range from {start_6} to {end_6}")

```
