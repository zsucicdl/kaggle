```python
import numpy as np
import pandas as pd
import pathlib
train_eegs = pathlib.Path("/kaggle/input/hms-harmful-brain-activity-classification/train_eegs")
train_specs = pathlib.Path("/kaggle/input/hms-harmful-brain-activity-classification/train_spectrograms")
```

## For all of your normalization needs
Contains means/variance for columns for eeg/spectrograms in a numpy array.

I've never used dask before so thought I might as well try, it probably isn't the fastest way of doing this, was pretty easy and simple to do though.

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
```

## EEGS

```python
import dask.dataframe as dd
df = dd.read_parquet(train_eegs/"*.parquet")
```

```python
df.mean(axis=0, skipna=True).compute().to_numpy()
```

```python
df.var(axis=0, skipna=True).compute().to_numpy()
```

## SPECS

```python
df_spec = dd.read_parquet(train_specs/"*.parquet")
```

```python
df_spec.mean(axis=0, skipna=True).compute().to_numpy()
```

```python
df_spec.var(axis=0, skipna=True).compute().to_numpy()
```
