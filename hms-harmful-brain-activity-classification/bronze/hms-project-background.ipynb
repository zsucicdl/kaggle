# Background



## What is EEG? 
*See ref 1 & 2*

Electroencephalography (EEG) uses non-invasive electrodes placed on the scalp and measures the current produced during excitation of groups of neurons. The weak electrical current that these groups of neurons produce is massively amplified, resulting in an EEG recording.

The waves that we see in an EEG recording are sinusoidal in nature, and they are normally measured from peak to trough. There are four main classifications for EEG brain wave patterns: 

1. Beta waves - more than 13Hz
2. Alpha waves - 8-13Hz
3. Theta waves - 4-8Hz
4. Delta waves - 0.5-4Hz

Depending on the state of the subject one waveform will be more dominant thatn the rest (ex: beta waves dominate during wakefulness, delta waves dominate during non-REM deep sleep). 


## EEG Patterns

For any of these patterns to be a bona fide pattern, they need to occur at least 6 times one after another.

### Periodic 
Periodic discharges with intervals between them

### Rhythmic 
Discharges with no breaks

### Spike and Wave (rare)
Tight spikes and a longer wave afterwards, with no intervals
