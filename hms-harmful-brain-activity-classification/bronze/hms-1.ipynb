```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if '.csv' in filename:
            print(os.path.join(dirname, filename))
            
            
warnings.filterwarnings('ignore')
```

```python
def plot_votes(df0, eeg_id=1628180742):
    df1 = df0.query(f'eeg_id == {eeg_id}')
    df1.drop(['eeg_id','eeg_sub_id','spectrogram_id', 'spectrogram_sub_id', 'label_id', 'patient_id', 'expert_consensus', 'spectrogram_label_offset_seconds'], axis=1, inplace=True)
    df1.set_index('eeg_label_offset_seconds', inplace=True)
    df1.plot.bar(title=f'eeg id = {eeg_id}', xlabel='eeg label offset seconds', ylabel='Votes')
    
```
