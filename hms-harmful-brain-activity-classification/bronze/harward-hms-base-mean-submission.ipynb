```python
import pandas as pd
test_df = pd.read_csv("/kaggle/input/hms-harmful-brain-activity-classification/test.csv")
sample_df = pd.read_csv("/kaggle/input/hms-harmful-brain-activity-classification/sample_submission.csv")

submission_df = pd.DataFrame(columns=sample_df.columns.values)
submission_df['eeg_id'] = test_df['eeg_id']
```

```python
# Copied from Chris Deotte's notebook
df = pd.read_csv('/kaggle/input/hms-harmful-brain-activity-classification/train.csv')
TARGETS = df.columns[-6:]
TARS = {'Seizure':0, 'LPD':1, 'GPD':2, 'LRDA':3, 'GRDA':4, 'Other':5}
TARS2 = {x:y for y,x in TARS.items()}

train = df.groupby('eeg_id')[['patient_id']].agg('first')

tmp = df.groupby('eeg_id')[TARGETS].agg('sum')
for t in TARGETS:
    train[t] = tmp[t].values
    
y_data = train[TARGETS].values
y_data = y_data / y_data.sum(axis=1,keepdims=True)
train[TARGETS] = y_data

tmp = df.groupby('eeg_id')[['expert_consensus']].agg('first')
train['target'] = tmp
train_mean = train[TARGETS].mean()

train_mean
```

```python
for col in TARGETS:
    submission_df[col] = train_mean[col]
    
submission_df.to_csv("submission.csv", index=False)
```
