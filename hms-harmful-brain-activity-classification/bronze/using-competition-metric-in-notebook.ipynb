# Remarks

```python
import pandas as pd
```

`metric.score` is destructive.

```python
solution = pd.DataFrame({'id': range(3), 'ham': [0, 0.5, 0.5], 'spam': [0.1, 0.5, 0.5], 'other': [0.9, 0, 0]})
submission = pd.DataFrame({'id': range(3), 'ham': [0.2, 0.3, 0.5], 'spam': [0.1, 0.5, 0.5], 'other': [0.7, 0.2, 0]})
print('before score call')
print(solution)
print('score(solution, submission)', metric.score(solution, submission, 'id'))
print('after score call')
print(solution)
```

The Kullback-Leibler divergence is not symmetric.

```python
solution = pd.DataFrame({'id': range(3), 'ham': [0, 0.5, 0.5], 'spam': [0.1, 0.5, 0.5], 'other': [0.9, 0, 0]})
submission = pd.DataFrame({'id': range(3), 'ham': [0.2, 0.3, 0.5], 'spam': [0.1, 0.5, 0.5], 'other': [0.7, 0.2, 0]})
print('score(solution, submission)', metric.score(solution.copy(), submission.copy(), 'id'))
print('score(submission, solution)', metric.score(submission.copy(), solution.copy(), 'id'))
```
