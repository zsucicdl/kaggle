**INTRODUCTION TO EEG (ELECTRO ENCEPHALOGRAM)**

EEG shows brain activity of a person , the results can be normal and abnormal as interpeted changes subject to subject. we basically see for 4 types of waves 
* alpha wave 
* beta wave
* gamma wave
* theta wave(found at age below of 5yrs)

some common terms used for marking - 
* F - frontal
* C - central
* T - Temproral
* P - parietal 
* O - occipital ,etc.





**HOW CAN WE USE EEG FINDINGS**





EEG is the most common method for recording brain signals in humans
because it is safe, affordable, easy to use, and because it has a high temporal
resolution (of the order of milliseconds)
