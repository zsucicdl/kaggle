```python
import pandas as pd
import numpy as np
import os

metadata = pd.read_csv('/kaggle/input/hms-harmful-brain-activity-classification/train.csv')
spectro_path = '/kaggle/input/hms-harmful-brain-activity-classification/train_spectrograms'

spids = np.unique(metadata['spectrogram_id'])

sp_summary = pd.DataFrame()
for spid in spids:
    record = {}
    sp_data = pd.read_parquet(os.path.join(spectro_path, str(spid) + '.parquet'))
    record['spectrogram_id'] = spid
    record['min_val'] = np.min(sp_data.iloc[:, 1:].values)
    record['max_val'] = np.max(sp_data.iloc[:, 1:].values)
    record['nbr_rows'] = sp_data.shape[0]
    record['nbr_nan_rows'] = len(np.unique(np.where(sp_data.isna())[0]))
    print(record)
    sp_summary = pd.concat([sp_summary, pd.DataFrame([record])], axis=0, ignore_index=True)

sp_summary.to_csv('sp_summary.csv')
```
